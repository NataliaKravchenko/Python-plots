```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Nat\Downloads\housing (1).data",
                      sep='\\s+')
df.columns = ['crime_rate','<25,000 sq.ft','non-ret_buz','riv_tract','nox_conc','num_rooms','age_built<1940',
              'dist_empl','acc_highway','prop_tax','pup-teach_ratio','black','low_pop','price']
```

```python
plt.style.use("ggplot")
fig, axes = plt.subplots(1, 1, figsize=(5, 5))
#plt.clf()
```

```python

axes.scatter(df['prop_tax'], df['price'], alpha=0.7, label='Pirce')

axes.scatter(df['prop_tax'], df['acc_highway'], alpha=0.7, label='Highway access')
axes.scatter(df['prop_tax'], df['low_pop'], alpha=0.7, label='Population lower status')

axes.set_xlabel('Property tax')
axes.set_title('Tax dependancy')
axes.legend()

fig
```
