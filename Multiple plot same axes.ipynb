```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Nat\Desktop\Python datasets\housing.data",
                      sep='\\s+')
df.columns = ['crime_rate','<25,000 sq.ft','non-ret_buz','riv_tract','nox_conc','num_rooms','age_built<1940',
              'dist_empl','acc_highway','prop_tax','pup-teach_ratio','black','low_pop','price']
```

```python
plt.style.use("ggplot")
fig, axes = plt.subplots(1, 1, figsize=(5, 5))
#plt.clf()
```

```python

axes.scatter(df['dist_empl'], df['price'], alpha=0.5, label='Price')
axes.scatter(df['dist_empl'], df['non-ret_buz'], alpha=0.5, label='Business')
axes.scatter(df['dist_empl'], df['acc_highway'], alpha=0.5, label='Highway')

axes.set_xlabel('Distanse')
axes.set_title('Facilities')
axes.legend()

fig
```
