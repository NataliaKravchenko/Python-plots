```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df=pd.read_csv(r"C:\Users\Nat\Downloads\housing (1).data",
                      sep='\\s+')
df.columns = ['crime_rate','<25,000 sq.ft','non-ret_buz','riv_tract','nox_conc','num_rooms','age_built<1940',
              'dist_empl','acc_highway','prop_tax','pup-teach_ratio','black','low_pop','price']
```

```python
tract = df[df['riv_tract'] == 0]
no_tract = df[df['riv_tract'] == 1]
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1, projection='3d')
line1 = axes.scatter(tract['price'], tract['dist_empl'], tract['age_built<1940'])
line2 = axes.scatter(no_tract['price'], no_tract['dist_empl'], no_tract['age_built<1940'])
axes.legend((line1, line2), ('tract', 'no_tract'))
axes.set_xlabel('price')
axes.set_ylabel('dist_empl')
axes.set_zlabel('age_built<1940')
```
