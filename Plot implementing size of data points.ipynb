```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Nat\Downloads\housing (1).data",
                      sep='\\s+')
df.columns = ['crime_rate','<25,000 sq.ft','non-ret_buz','riv_tract','nox_conc','num_rooms','age_built<1940',
              'dist_empl','acc_highway','prop_tax','pup-teach_ratio','black','low_pop','price']
```

```python
plt.style.use("ggplot")
fig, axes = plt.subplots(1, 1, figsize=(8, 8))
#plt.clf()
```

```python
axes.scatter(df['age_built<1940'], df['dist_empl'], s=(df['price']) ** 1.5,
             label=f'House quality', color='green', marker='o', edgecolors='w', alpha=0.7)

axes.set_xlabel(f'built<1940')
axes.set_ylabel(f'dist_empl')
axes.set_title(f'Units built prior to 1940, Distance to employment and Price (size)')
axes.legend()
fig
```
