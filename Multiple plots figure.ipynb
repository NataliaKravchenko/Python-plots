```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Nat\Desktop\Python datasets\housing.data",
                      sep='\\s+')
df.columns = ['crime_rate','<25,000 sq.ft','non-ret_buz','riv_tract','nox_conc','num_rooms','age_built<1940',
              'dist_empl','acc_highway','prop_tax','pup-teach_ratio','black','low_pop','price']
```

```python
fig, axes = plt.subplots(3, 2, figsize=(8,8))
fig.tight_layout(pad=4.0)

axes[0][0].plot(df['price'], color='b')
axes[0][0].set_title('Property price')
axes[0][0].set_ylabel('Property price')

axes[0][1].plot(df['prop_tax'], color='b')
axes[0][1].set_title('Property tax')
axes[0][1].set_ylabel('Property tax')

axes[1][0].scatter(df['prop_tax'], df['crime_rate'], color='y')
axes[1][0].set_title('Tax vs crime rate')
axes[1][0].set_xlabel('Property tax')
axes[1][0].set_ylabel('Crime rate')

axes[1][1].scatter(df['prop_tax'], df['non-ret_buz'], color='y')
axes[1][1].set_title('Tax vs business activity')
axes[1][1].set_xlabel('Property tax')
axes[1][1].set_ylabel('Non retail business')

axes[2][0].scatter(df['prop_tax'], df['nox_conc'], color='g')
axes[2][0].set_title('Tax vs nox concentration')
axes[2][0].set_xlabel('Property tax')
axes[2][0].set_ylabel('nox concentration')

axes[2][1].scatter(df['prop_tax'], df['acc_highway'], color='g')
axes[2][1].set_title('Tax vs highway access')
axes[2][1].set_xlabel('Property tax')
axes[2][1].set_ylabel('Highway access')

```
