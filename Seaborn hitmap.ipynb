```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Nat\Desktop\Python datasets\housing.data",
                      sep='\\s+')

df.columns = ['crime_rate','<25,000 sq.ft','non-ret_buz','riv_tract','nox_conc','num_rooms','age_built<1940',
              'dist_empl','acc_highway','prop_tax','pup-teach_ratio','black','low_pop','price']


fig, axes = plt.subplots(figsize=(12,10))
sns.heatmap(df.corr(), annot=True, cmap='winter')
axes.set_xticklabels(df.columns, rotation=90)
axes.set_yticklabels(df.columns)
```
